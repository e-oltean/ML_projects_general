```python
import numpy
import matplotlib.pyplot as plt

x = [2,3,4,5,6,8,9,11,12,13,14,16,19,20,21,23,25,27]
y = [200,180,170,70,60,55,60,45,55,60,65,70,80,90,100,180,190,200]

mymodel = numpy.poly1d(numpy.polyfit(x, y, 6))

myline = numpy.linspace(1, 22, 200)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show() 
```
